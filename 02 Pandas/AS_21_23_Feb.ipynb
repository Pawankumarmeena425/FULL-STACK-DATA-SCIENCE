```python
# Q7. Create a DataFrame using multiple Series. Explain with an example.

# create a Series for names
names = pd.Series(['Alice', 'Bob', 'Charlie', 'David'])

# create a Series for ages
ages = pd.Series([25, 32, 18, 47])

# create a Series for cities
cities = pd.Series(['New York', 'Paris', 'London', 'Tokyo'])

# create a DataFrame by combining the Series
df = pd.concat([names, ages, cities], axis=1)

# set column names for the DataFrame
df.columns = ['Name', 'Age', 'City']

# display the DataFrame
print(df)
```

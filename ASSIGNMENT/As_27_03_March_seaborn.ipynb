```python
import seaborn as sns
```

```python
# Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.

# 1. Bar Plot: A bar plot displays categorical data with rectangular bars whose lengths represent the values of the data. 
# 2. Scatter Plot: A scatter plot displays the relationship between two continuous variables. It uses dots to represent individual data points
# 3. Box Plot: A box plot (also known as a whisker plot) displays the distribution of a continuous variable across different categories or groups. 
# 4. Histogram: A histogram is a graphical representation of the distribution of a continuous variable.
# 5. Heatmap: A heatmap visualizes the magnitude of a numerical variable in a two-dimensional grid. 
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using
# x ="timepoint" and y = "signal" for different events and regions.

fmri = sns.load_dataset('fmri')
fmri
```

```python
# graph plot betweeen x = "timepoint" and y = "signal" for 'event'
sns.scatterplot(x = fmri.timepoint , y = fmri.signal , data = fmri , hue = 'event')
```

```python
# graph plot betweeen x = "timepoint" and y = "signal" for 'region'
sns.scatterplot(x = fmri.timepoint , y = fmri.signal , data = fmri , hue = 'region')
```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn.
# Plot two box plots using x ='pclass', y = 'age' and y = 'fare'.
titanic = sns.load_dataset('titanic')
titanic
```

```python
sns.boxplot(x = titanic.pclass , y = titanic.age)
```

```python
# 1st way
sns.boxplot(x = titanic.pclass , y = titanic.age , width = 0.3)
```

```python
# 2nd way
sns.boxplot(x = titanic['pclass'] , y = titanic['fare'] ,color = 'r')
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue parameter for the 'cut' column of the diamonds dataset.
diamond = sns.load_dataset('diamonds')
diamond
```

```python
sns.displot(diamond.price)
```

```python
diamond['price'].value_counts()
```

```python
sns.relplot(x= diamond.price, y = diamond.carat , data = diamond , hue = 'cut')
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column of the iris dataset.
iris = sns.load_dataset('iris')
iris
```

```python
sns.relplot(x = iris.sepal_length , y = iris.petal_width , data = iris , hue = 'species')
```

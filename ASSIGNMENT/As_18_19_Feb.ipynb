```python
from flask import Flask

app = Flask(__name__)

@app.route("/hello")
def hello_world():
    return "<h1>Hello, World!</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0",port=3000)
```

```python
# Q3. What is App routing in Flask? Why do we use app routes?

''' App Routing : app routing refers to the process of mapping URLs or routes to specific functions in the Flask application. 
When a user makes a request to a specific URL, Flask matches the requested URL ,  and then go to the specific method through the binding routing .'''

# We use app routes in Flask for the mentioned reasons : 
# 1. Routing allows us to define different URL patterns for different parts of our application.through the routing we would be able to access the desire function or code in the application.
# 2 . Routing allows us to handle different HTTP methods (GET, POST, PUT, DELETE, etc.) for different URLs. 
# 3 . Routing allows us to pass variables or parameters in the URL, which can be useful for creating dynamic web applications.

```

```python
# Q4. Create a “/welcome” route to display the welcome message “Welcome to ABC Corporation” and a “/”  route to show the following details:
# Company Name: ABC Corporation
# Location: India
# Contact Detail: 999-999-9999

from flask import Flask

app = Flask(__name__)

@app.route("/")
def information():
    return '<p> companyName: "ABC Corparation",<br> Location: "India",<br> Contact :"999-999-9999"</p>'

@app.route("/welcome")
def welcome_msg():
    return "<h1>Welcome to the ABC Corparation !</h1>"

if __name__=="__main__":
    app.run(host="0.0.0.0",port=3000)
```

```python
# Q5. What function is used in Flask for URL Building? Write a Python code to demonstrate the working of the url_for() function.

# In Flask, the url_for function is used for URL building.
# code:

from flask import Flask,url_for

app = Flask(__name__)

@app.route("/")
def information():
    data = {
    "companyName" : "ABC Corparation",
    "Location": "India",
    "Contact" :"999-999-9999"
    }
    return data

@app.route("/welcome")
def welcome_msg():
    return "<h1>Welcome to the ABC Corparation !</h1>"

print("Use of url_for function for print url ")
with app.test_request_context():
    print(url_for('information'))
    print(url_for('welcome_msg'))
    
    
if __name__=="__main__":
    app.run(host="0.0.0.0",port=3000)
```

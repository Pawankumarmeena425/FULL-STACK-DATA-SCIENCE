```python
# Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
# Note: Also add title, xlabel, and ylabel to the plot.
```

```python
# Scatter Plot: A scatter plot is used to investigate the relationship between two continuous variables. It displays individual data points as markers on a Cartesian plane, where the position of each point represents the values of the two variables.
plt.scatter(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Scatter plot")
```

```python
# subplot() : The subplot() function in Matplotlib is used to create a grid of subplots within a single figure. It allows you to display multiple plots or charts together, making it easier to compare and analyze different aspects of your data
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the data for each line plot
x = np.array([0, 1, 2, 3, 4, 5])

y1 = np.array([0, 100, 200, 300, 400, 500])
y2 = np.array([50, 20, 40, 20, 60, 70])
y3 = np.array([10, 20, 30, 40, 50, 60])
y4 = np.array([200, 350, 250, 550, 450, 150])

# Create a figure with four subplots
fig, axs = plt.subplots(2,2, figsize=(10, 8))

# Plotting the first line plot
axs[0, 0].plot(x, y1, color='red')
axs[0, 0].set_title('Line Plot 1')

# Plotting the second line plot
axs[0, 1].plot(x, y2, color='blue')
axs[0, 1].set_title('Line Plot 2')

# Plotting the third line plot
axs[1, 0].plot(x, y3, color='green')
axs[1, 0].set_title('Line Plot 3')

# Plotting the fourth line plot
axs[1, 1].plot(x, y4, color='orange')
axs[1, 1].set_title('Line Plot 4')

# Adjust the spacing between subplots
plt.tight_layout()

# Display the figure
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# bar plot : A bar plot is a type of chart that uses rectangular bars to represent categorical data.
#  It is commonly used to compare and visualize data across different categories or groups.
```

```python
plt.bar(company ,  profit , color ='r')
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Company Profits')
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
# box plot : A box plot, also known as a box-and-whisker plot, is a statistical visualization that provides a summary of the distribution of a dataset. It displays key statistical measures, including the minimum, maximum, median, and quartiles, as well as any potential outliers.

# Box plots are used for the following purposes:

# Summary of Data Distribution
# Identification of Outliers
# Comparison of Groups
# Visualization of Skewness
```

```python
# Combine the data into a list
data = [box1, box2]
# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Box 1', 'Box 2'])
plt.xlabel('Box')
plt.ylabel('Value')
plt.title('Box Plot')
plt.show()
```
